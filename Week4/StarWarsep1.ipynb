```python
# Import packages
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import json
from fa2 import ForceAtlas2
```

```python
# ### Creating a graph
#
# Create an empty graph with no nodes and no edges.

# By definition, a `Graph` is a collection of nodes (vertices) along with
# identified pairs of nodes (called edges, links, etc).

G = nx.Graph()
```

```python
#read json files
#Add nodes
#Add edges


with open('starwars-episode-1-interactions.json') as f:
    json_data = json.loads(f.read())
#Nodes - Name,value,color
#Links - source ,target,value


json_data['nodes']
#Load json data
#Add nodes based on the nodes name
#For Links,
#Loop through the links add edges from source to target
#The value will be the weight
```

```python
len(json_data['nodes'])
name_nodes= []

for i in range(len(json_data['nodes'])):
    #print(json_data['nodes'][i]['name'])
    name_nodes.append(json_data['nodes'][i]['name'])

print(name_nodes)
print(len(name_nodes))
```

```python
G.add_nodes_from(name_nodes)
```

```python
json_data['links']
```

```python
for i in range (len(json_data['links'])):
    G.add_edge(name_nodes[json_data['links'][i]['source']],name_nodes[json_data['links'][i]['target']],weight = json_data['links'][i]['value'])
```
