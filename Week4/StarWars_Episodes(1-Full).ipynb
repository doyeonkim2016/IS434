```python
#Cracking Star Wars
#KDY
#Objective: We will be only focusing on the given interaction of the characters
#Degree Centrality - What is tha main character of the episode?
#Betwenness- How integral it is in the episode
#Closeness - How close it is in the episode
#import pakages
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import json

```

```python
# ### Creating a graph...
G = nx.Graph()
```

```python
#Nodes - Name,value,color
#Links - source ,target,value
filename='starwars-full-interactions.json'
with open(filename) as f:
    json_data = json.loads(f.read())


json_data['nodes']
```

```python
len(json_data['nodes'])
name_nodes= []

#Add Nodes
for i in range(len(json_data['nodes'])):
    name_nodes.append(json_data['nodes'][i]['name'])

#Add Edges    
for i in range (len(json_data['links'])):
    G.add_edge(name_nodes[json_data['links'][i]['source']],name_nodes[json_data['links'][i]['target']],weight = json_data['links'][i]['value'])
```

```python
nx.draw(G,with_labels=True)
nx.write_gexf(G, filename[0:-5]+".gexf")
```

```python
print("Degree Histogram: ",nx.degree_histogram(G))
print("Nodes are:\n\t", list(G.nodes) )
print("Edges are:\n\t", list(G.edges) )
print("Stat info: ", nx.info(G))
```
